```python
import pandas as pd
import numpy as np
import os
```

```python
path = 'names'
cols = ['Name', 'Gender', 'Count']
```

```python
def count_top3(years):
    names_years = []
    for year in years:
        tmp_names = pd.read_csv(os.path.join(path, 'yob' + year + '.txt'),
                               names=cols)
        names_years.append(tmp_names)
        print(tmp_names)
        
    if len(names_years) > 1:
        all_names = pd.merge(names_years[0], names_years[1], on=['Gender'])
        
        if len(names_years) > 2:
            for i in range(2, len(names_years)):
                all_names = pd.merge(
                all_names, names_years[i], on=['Name', 'Gender', 'Count'])

        print(all_names)
```

```python
count_top3(['1995', '2000'])
```
